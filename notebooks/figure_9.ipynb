# Figure 9. MRVL of error distributions

### Imports

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from data import load_data
from vis import plot_mrvl_by_response, set_export_style, SET_SIZE_PALETTE

save_fig = False
```

### Figure 9A (upper). Continuous MRVL free report order

```python
# MRVL by response (continuous, free)
modality = 'color'
dataset = 'continuous'
condition = 'free'

df = load_data(modality=modality, condition=condition, dataset=dataset)
fig, ax = plot_mrvl_by_response(df)

set_export_style(fig, [ax], figsize=(6, 4))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_position(('outward', 10))
ax.spines['bottom'].set_bounds((1, 6))
ax.set_ylim([0, 1])
ax.set_yticks([0, 0.5, 1])
ax.set_yticklabels(['0', '0.5', '1'])
ax.get_legend().remove()

if save_fig:
    plt.savefig('../figures/fig_S02-mrvl_cont_free.png', bbox_inches='tight', dpi=800)

plt.show()
```
