# Figure 1. Whole-report task comparison

### Imports

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from data import load_data
from vis import DISC_EDGES, plot_error_by_set_size, plot_error_by_response, plot_error_by_response_stacked, plot_error_by_response_ecdf, set_axes_labels, set_export_style, plot_paired_error_by_set_size, plot_paired_error_by_response, set_axes_aspect

save_fig = False
```

### Load data

```python
### DATASET LOADING ###
modality = 'color'
set_size = 6 # for response plotting

# continuous datasets
df_free_cont = load_data(modality=modality, condition='free', dataset='continuous')
df_free_cont_6 = df_free_cont[df_free_cont.set_size == set_size].reset_index(drop=True)

df_rand_cont = load_data(modality=modality, condition='rand', dataset='continuous')
df_rand_cont_6 = df_rand_cont[df_rand_cont.set_size == set_size].reset_index(drop=True)

# discrete datasets
df_free_disc = load_data(modality=modality, condition='free', dataset='discrete')
df_free_disc_6 = df_free_disc[df_free_disc.set_size == set_size].reset_index(drop=True)

df_rand_disc = load_data(modality=modality, condition='rand', dataset='discrete')
df_rand_disc_6 = df_rand_disc[df_rand_disc.set_size == set_size].reset_index(drop=True)
```

### Figure 1C. Continuous error distributions

```python
## errors by set size (continuous)##
fig, axes = plot_paired_error_by_set_size(df_free_cont, df_rand_cont, bins=90)

for ax, ss in zip(axes[1:], [2, 3, 4, 6]):
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_xticks([])
    ax.set_xticklabels([])
    ax.set_yticks([])
    ax.set_ylim(-.15,.15)

axes[0].set_yticks([-.15, 0, .15])
axes[0].set_yticklabels([0.15, 0, 0.15])
axes[0].spines['left'].set_position(('outward',10))
axes[0].spines['bottom'].set_position(('outward',10))
axes[0].spines['right'].set_visible(False)
axes[0].spines['top'].set_visible(False)
axes[0].set_ylim(-.15, .15)

set_axes_aspect(fig, axes)
set_export_style(fig, axes, remove_ticklabels=False, figsize=(10, 3))

if save_fig:
    plt.savefig('../figures/fig_01-02_cumulative_error_continuous.png', bbox_inches='tight', dpi=800)

plt.show()
```

```python
## errors by response (continuous)
fig, axes = plot_paired_error_by_response(df_free_cont_6, df_rand_cont_6, bins=90)

for ax, ss in zip(axes[1:], [2, 3, 4, 6, 8]):
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_xticks([])
    ax.set_xticklabels([])
    ax.set_yticks([])
    ax.set_ylim(-.1,.1)

axes[0].set_yticks([-.1, 0, .1])
axes[0].set_yticklabels([.1, 0, .1])
axes[0].spines['left'].set_position(('outward',10))
axes[0].spines['bottom'].set_position(('outward',10))
axes[0].spines['right'].set_visible(False)
axes[0].spines['top'].set_visible(False)
axes[0].set_ylim(-.1, .1)

set_axes_aspect(fig, axes)
set_export_style(fig, axes, remove_ticklabels=False, figsize=(10, 3))

plt.subplots_adjust(wspace=0.1)

if save_fig:
    plt.savefig('../figures/fig_01-04_response_error_continuous.png', bbox_inches='tight', dpi=800)

plt.show()
```
